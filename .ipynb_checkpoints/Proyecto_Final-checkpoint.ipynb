## De acuerdo al diagrama y al analisis previo, no se utilizan los siguientes atributos:

- eval_set de la tabla df_orders: Este atributo indica a cuál conjunto de datos (entrenamiento, prueba, etc.) pertenece cada pedido, pero no se usa en este análisis.
- days_since_prior_order de la tabla df_orders: Este atributo indica cuántos días han pasado desde el último pedido del mismo usuario, pero no se usa en este análisis.

## **4. Identificar los productos que más se compran juntos, es decir, aquellos que aparecen en los mismos pedidos con mayor frecuencia, y visualizar los resultados utilizando un gráfico de red.**

## 1. Visualización

## 1.1 Visualización de los productos que más se compran juntos.

```python
from itertools import combinations
from collections import Counter

# Agrupar los datos por 'order_id' y obtener la lista de productos en cada pedido
order_products = data.groupby('order_id')['product_name'].apply(list)

# Generar todas las posibles combinaciones de productos en cada pedido y contar la frecuencia de cada combinación
product_combinations = Counter()

for products in order_products:
    product_combinations.update(combinations(products, 2))

# Obtener las 10 combinaciones de productos más comunes
most_common_combinations = product_combinations.most_common(10)

# Imprimir las 10 combinaciones de productos más comunes
for (product1, product2), count in most_common_combinations:
    print(f'{product1} y {product2} se compraron juntos {count} veces')
```

```python
# Creamos un grafo
G = nx.Graph()

# Añadimos nodos (productos) y bordes (conexiones entre productos) al grafo
for (p1, p2), count in most_common_combinations:
    G.add_node(p1)
    G.add_node(p2)
    G.add_edge(p1, p2, weight=count)

# Dibujamos el grafo
plt.figure(figsize=(10,10))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=2000, font_size=10)
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.title('Los 10 productos más comprados juntos')
plt.show()
```

<small> Este código está utilizando la biblioteca NetworkX para crear un gráfico de red que muestra las combinaciones más comunes de productos comprados juntos. Aquí te explico cada paso:

G = nx.Graph(): Este comando crea un nuevo objeto grafo vacío, que puede llenarse con nodos y bordes.

for (p1, p2), count in most_common_combinations:: Esto es un bucle for que itera a través de la lista most_common_combinations, que asumimos es una lista de tuples, donde cada tuple tiene dos elementos: otro tuple con dos productos (p1, p2) y un conteo count de cuántas veces se compraron juntos esos dos productos.

G.add_node(p1), G.add_node(p2): Estos comandos añaden nodos al gráfico. Cada nodo representa un producto. Si el producto ya se ha añadido al grafo como nodo en una iteración anterior del bucle, estos comandos no tienen efecto.

G.add_edge(p1, p2, weight=count): Esto añade un borde entre los dos productos/nodos p1 y p2. El "peso" del borde es la cantidad de veces que esos dos productos se compraron juntos, representado por count.

plt.figure(figsize=(10,10)): Esto crea una nueva figura en la que se trazará el grafo.

pos = nx.spring_layout(G): Esto genera las posiciones de los nodos para el trazado del grafo. El diseño de "muelle" trata de posicionar los nodos de manera que todos los bordes sean más o menos igual de largos y haya tan pocas intersecciones de bordes como sea posible.

nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=2000, font_size=10): Esto dibuja el grafo. Los nodos son de color azul cielo, de tamaño 2000, y se dibujan con etiquetas (los nombres de los productos). Las posiciones de los nodos se determinan por pos.

labels = nx.get_edge_attributes(G, 'weight'): Esto recupera las etiquetas de los bordes, que son los pesos de los bordes.

nx.draw_networkx_edge_labels(G, pos, edge_labels=labels): Esto dibuja las etiquetas de los bordes, que son las cantidades de veces que se compraron juntos los dos productos conectados por cada borde.

plt.title('Los 10 productos más comprados juntos'): Esto añade un título al gráfico.

plt.show(): Esto muestra el gráfico en pantalla.
