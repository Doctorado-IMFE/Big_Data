### Se van a juntar todos los datos en un solo data frame que se llama:
- **data**

```python
 data = df_orders.merge(df_order_products, on="order_id", how="right").merge(df_products, on="product_id", how="left").merge(df_department, on="department_id", how="left").merge(df_aisles, on="aisle_id", how="left")
```

### Se muestra los primeros datos de **data**

```python
data.head()
```

### En este paso con el **data** se hara el siguiente proceso. 
1. La cantidad de productos por pedido.

```python
# Calcular la cantidad de productos por pedido
cant_prod_pedido = data.groupby('order_id')['product_id'].count()
"""1.Se agruparon los datos por pedido. 2.Con la instruccion ['product_id'].count() se cuenta el número de productos en cada pedido. 3.Se guarda este set de datos en la variable cant_prod_pedido"""
```

```python
# Crear un histograma de la cantidad de productos por pedido
plt.figure(figsize=(10,6))
plt.hist(cant_prod_pedido, bins=30, edgecolor='black')
plt.title('Distribución de la cantidad de productos por pedido')
plt.xlabel('Cantidad de productos')
plt.ylabel('Cantidad de pedidos')
plt.show()
```

2. La frecuencia de compras por día.

```python
# Calcular la frecuencia de compras por día de la semana
order_frequency = data['order_dow'].value_counts().sort_index()
""" Primero, vamos a contar cuántos pedidos se hicieron en cada día de la semana. Esto se puede hacer con la función value_counts(). Luego, vamos a ordenar los resultados por el día de la semana (de 0 a 6) usando la función sort_index().

Después, vamos a graficar los resultados en un gráfico de barras con la función plot.bar().

value_counts().sort_index() cuenta el número de veces que aparece cada valor en la columna 'order_dow' y luego ordena los resultados por el índice (que es el día de la semana en este caso).
"""
```

```python
# Crear un gráfico de barras de la frecuencia de compras por día de la semana
order_frequency.plot.bar(figsize=(10,6))
plt.title('Frecuencia de compras por día de la semana')
plt.xlabel('Día de la semana')
plt.ylabel('Cantidad de pedidos')
plt.show()
"""plot.bar() crea un gráfico de barras de los datos.
plt.title(), plt.xlabel(), y plt.ylabel() añaden un título al gráfico y etiquetas a los ejes x e y, respectivamente.
plt.show() muestra el gráfico."""
```

3. La frecuencia de compras por hora.

```python
# Calcular la frecuencia de compras por hora del día
order_frequency_hour = data['order_hour_of_day'].value_counts().sort_index()
""" Similar al caso anterior, vamos a contar cuántos pedidos se hicieron en cada hora del día usando la función value_counts(). Luego, vamos a ordenar los resultados por la hora del día (de 0 a 23) usando la función sort_index(). 
value_counts().sort_index() cuenta el número de veces que aparece cada valor en la columna 'order_hour_of_day' y luego ordena los resultados por el índice (que es la hora del día en este caso)."""
```

```python
# Crear un gráfico de barras de la frecuencia de compras por hora del día
order_frequency_hour.plot.bar(figsize=(10,6))
plt.title('Frecuencia de compras por hora del día')
plt.xlabel('Hora del día')
plt.ylabel('Cantidad de pedidos')
plt.xticks(range(24))
plt.show()
""" plot.bar() crea un gráfico de barras de los datos.
plt.title(), plt.xlabel(), y plt.ylabel() añaden un título al gráfico y etiquetas a los ejes x e y, respectivamente.
plt.xticks(range(24)) ajusta las marcas en el eje x para que muestren cada hora del día.
plt.show() muestra el gráfico. """
```

## 4. **Identificar los productos más comprados en cada departamento y visualizar los resultados utilizando un gráfico de barras.**

```python


# Agrupa por 'department' y 'product_name', cuenta la cantidad de pedidos para cada producto
product_counts = data.groupby(['department', 'product_name']).size().reset_index(name='counts')

# Para cada departamento, encuentra el producto con la mayor cantidad de pedidos
top_products = product_counts.loc[product_counts.groupby('department')['counts'].idxmax()]

# Ordena los datos por la cantidad de pedidos
top_products_sorted = top_products.sort_values(by='counts', ascending=False)

# Crea un gráfico de barras horizontal
plt.figure(figsize=(10,8))
sns.barplot(y='department', x='counts', data=top_products_sorted, orient='h', palette='viridis')

# Añade el nombre del producto más vendido como etiqueta en cada barra
for i in range(top_products_sorted.shape[0]):
    plt.text(top_products_sorted.counts.iloc[i], i,
             top_products_sorted.product_name.iloc[i],
             va='center')

plt.title('Producto más comprado por departamento')
plt.xlabel('Cantidad de pedidos')
plt.ylabel('Departamento')
plt.show()

"""
Este código crea un gráfico de barras horizontal donde cada barra representa un departamento y su longitud corresponde a la cantidad de pedidos del producto más comprado en ese departamento. Adicionalmente, el nombre del producto más comprado se muestra como etiqueta en cada barra.
"""
```

## **3. Identificar los productos que más se compran juntos, es decir, aquellos que aparecen en los mismos pedidos con mayor frecuencia, y visualizar los resultados utilizando un gráfico de red.**

```python
!pip install networkx

import networkx as nx
from itertools import combinations
from collections import Counter

# Creamos una lista vacía para almacenar las combinaciones de productos en cada pedido
product_combinations = []

# Agrupamos los datos por 'order_id' y 'product_name'
for order_id, order_df in data.groupby('order_id'):
    # Creamos las combinaciones de productos (2 productos a la vez) en cada pedido
    product_combinations += list(combinations(order_df['product_name'].values, 2))

# Contamos la cantidad de veces que cada combinación de productos ocurre
product_combination_counts = Counter(product_combinations)

# Obtenemos las 10 combinaciones de productos más comunes
most_common_combinations = product_combination_counts.most_common(10)

# Creamos un grafo
G = nx.Graph()

# Añadimos nodos (productos) y bordes (conexiones entre productos) al grafo
for (p1, p2), count in most_common_combinations:
    G.add_node(p1)
    G.add_node(p2)
    G.add_edge(p1, p2, weight=count)

# Dibujamos el grafo
plt.figure(figsize=(10,10))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=2000, font_size=10)
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.title('Los 10 productos más comprados juntos')
plt.show()
```
